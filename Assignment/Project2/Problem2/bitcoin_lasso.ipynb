```python
import pandas as pd
bitcoin=pd.read_csv('/Users/feng/Desktop/bitcoin_dataset.csv')
bitcoin.head()
```

```python
bitcoin=bitcoin.drop(['Date','btc_trade_volume'],axis=1)
bitcoin.info()
```

```python
bitcoin.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
bitcoin.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
import numpy as np
def split_train_test(data):
    indices=np.arange(len(data))
    test_indices=indices[2645:]
    train_indices=indices[:2645]
    return data.iloc[train_indices],data.iloc[test_indices]
train_set,test_set=split_train_test(bitcoin)
print(len(train_set))
```

```python
train_labels=train_set["btc_market_price"].copy()
train_features=train_set.drop("btc_market_price",axis=1)
test_labels=train_set["btc_market_price"].copy()
test_features=train_set.drop("btc_market_price",axis=1)
```

```python
#alpha and error
from sklearn.linear_model import Lasso
import numpy as np
def lasso_compare(alphas,train_labels,train_features,test_labels,test_features):
    errors=[]
    for alpha in alphas:
        clf=Lasso(alpha)
        clf.fit(train_features,train_labels)
        prediction=clf.predict(test_features)
        avg_square_error=np.mean(np.square(test_labels-prediction))
        errors.append(avg_square_error)
    return errors

alphas=np.hstack((np.arange(0.01,0.1,0.01),np.arange(0.1,1,0.1)))
errors=lasso_compare(alphas,train_labels,train_features,test_labels,test_features)

```

```python
#alpha:0.01-0.1,0.1-0.9
plt.scatter(alphas,errors)
plt.show()
```

```python
#alpha:0.01-0.1
alphas=np.arange(0.001,0.01,0.001)
errors=lasso_compare(alphas,train_labels,train_features,test_labels,test_features)
plt.scatter(alphas,errors)
plt.show()
```

```python
#alpha below 0.001 , error doesn't shrink anymore.So i set alpha=0.001 finally.
alphas=np.arange(0.0001,0.001,0.0001)
errors=lasso_compare(alphas,train_labels,train_features,test_labels,test_features)
plt.scatter(alphas,errors)
plt.show()
```

```python
from sklearn.externals import joblib
clf=Lasso(alpha=0.001)
clf.fit(train_features,train_labels)
joblib.dump(clf, "train_model.m")
```
